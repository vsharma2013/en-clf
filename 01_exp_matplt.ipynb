```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys, os, re, json, time
from datetime import datetime
import yfinance as yf
```

```python
m = 30
n = 6

data = np.zeros((m,n))

for i in range(n):
    data[:, i] = 30*np.random.randn(m)*(2*i/(n-1) - 1)**2 + (i+1)**2
```

```python
fig, ax = plt.subplots(1, 3, figsize=(8, 2))

ax[0].bar(np.arange(n), data.mean(axis=0))
ax[0].set_title('Bar Plot')

ax[1].errorbar(np.arange(n), data.mean(axis=0), np.std(data, axis=0, ddof=1), marker='s', linestyle='')
ax[1].set_title('Error Plot')

ax[2].bar(np.arange(n), data.mean(axis=0))
ax[2].errorbar(np.arange(n), data.mean(axis=0), np.std(data, axis=0, ddof=1), marker='s', linestyle='', color='red')
ax[2].set_title('Bar+Err Plot')
```

```python
m = [[2,5,4,3],[1,1,8,6]]

fig, ax = plt.subplots(2,2, figsize = (8,8))

ax[0,0].imshow(m)

df = pd.DataFrame(m, columns='c1 c2 c3 c4'.split())
df.plot(ax=ax[1,0], kind='bar')
ax[1,0].set_title('Group by rows')

ax[0,1].imshow(np.array(m).T)
df.T.plot(ax=ax[1,1], kind='bar')
ax[1,1].set_title('Group by columns')
```
