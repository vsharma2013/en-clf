```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.pairplot(tips, hue='sex')
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, hue='size')
```

```python
sns.countplot(x='size', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', split=True, jitter=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
sns.heatmap(tips.corr(), annot=True)
```

```python
sns.heatmap(flights.corr(), annot=True)
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.clustermap(fp, cmap='coolwarm')
```
